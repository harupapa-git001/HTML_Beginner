ボックスモデル

ここで、css側面からhtmlを見てみましょう。
htmlはボックスモデルという考え方でcssによって装飾していきます。ボックスモデルとは、つまり「箱」あるいは「四角形」として考えるということです。どういうことか、詳しくみていきます。
ほとんど全てのタグは下図のような4重構造の四角形でできています。

内側から中身、padding, border, marginのようになっています。
paddingは内側の余白です。先ほどh1タグで中身の左に余白を作りましたね。
marginは外側の余白です。
borderは外側と内側の中間に位置し、境界を意味しています。
このボックスモデルを人体で例えるなら、他者との距離(ソーシャルディスタンス)がmarginで、borderが皮膚、paddingが筋肉や皮下脂肪で中身が骨となります。
ちなみに、background-colorはpaddingより内側に色をつけます。
いまいち意味がまだ理解できないと思うので、実際にcssを書いて説明していきます。
まずは違いがややこしいpaddingとmarginの違いを見ていきましょう。

paddingとmarginの違い

内側の余白paddingと外側の余白marginは違いや使い所の見極めが難しく、初心者を苦しめてきました。
この違いを見るために、背景色のついているh1へもう一度登場してもらいます。

「初めてのブログ投稿」というのが人体で言うところの骨格です。先ほどは、左側にpaddingを設定しました。
今度は、上側にpaddingを設定してみましょう。

見比べてみると、文字の上に少しスペースが空いたのがわかるかと思います。
次に、上側にmarginを設定してみます。分かりやすいように、今だけ画面全体(body)へ背景色も設定しておきます。

marginもpaddingと同じように、4方向top, bottom, left, rightのそれぞれを指定することができます。

青色のh1の上にスペースが空くようになりました。
paddingやborderを含めた塊が、上から距離を取ったと言う感じです。
同じように、投稿日時とも距離をとりましょう。

h1と投稿日時が離れたのがわかるかと思います。
このように、他のタグ(要素)と距離をとりたい時はmarginを使い、自分自身の中に余裕を持ちたい時はpaddingを使います。
このとき、h1の下にmarginをつけるのか、投稿日時の上にmarginをつけるのか、結果的には同じ見た目になりますが、どちらを選ぶかは正直好みの問題かなと思います。
ここは実装する人によってもやり方は分かれます。

border

最後に、paddingとmarginの間に存在するborderを見ていきます。borderも同じようにtop, bottom, left, rightそれぞれにつけることができますが、少し違うやり方も見ていきましょう。

borderには渡すものがいろいろあります。solidは線の種類の１つで、直線を表します。点線などを選ぶこともできます。
3pxは線の太さ、最後のredは線の色です。

色を変えてみたり、3pxを5pxに変えてみたり、いろいろと実験してみると、どこがどのように作用しているのかが分かり易いです。
borderの後ろに「-top」や「-right」をつけないと、上下左右全てを同時に指定したことになります。
同じように、「margin」や「padding」も単体で設定することができます。

「padding: 20px」と書くと、上下左右それぞれに20pxの内側の余白を設定したことになります。

blockとinline

ボックスモデルを使いこなす上で避けて通れないのが、blockとinline(と中間のinline-block)の2つです。
blockとinlineはタグごとに設定してあるスタイルです。

この1行占領する方をblock要素、1行占領しない方をinline要素と呼んでいます。
