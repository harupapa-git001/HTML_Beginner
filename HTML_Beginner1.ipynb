タグの意味を理解しよう

HTMLにおいて最も重要なのが「タグ」です。タグとは目印のようなもので、特定の意味を持ちます。タグの種類はたくさんなるのですが、それぞれの持つ意味についてはこれから一つずつ順番にみていきます。

```python
<h1>出版社</h1>
```

のように

```python
<タグ名>中身</タグ名>
```

終了タグ(後ろのタグ)がある場合は「</タグ名>」のように「/」を必ずつけます。終了タグがないものもありますが、ひとまずこの形が基本形です。タグの中にはまたタグを入れることができます。

```python
<div>
    <h1>ミナトミ出版社</h1>
</div>
```

この場合はdiv(ディブ)タグの中にh1(エイチワン)タグが入っています。HTMLでは、「**の中にxxが入る　＝入れ子構造」によって、データの「木構造」をうまく表現しているのです。難しいことは考えず、今つのポイント

・開始タグと終了タグで囲む

・タグの中にもタグを入れることができる

だけ抑えておいてください。
ではここからよく使うタグを１つずつ見ていきましょう。

h数字

h1やh2などがあり、h6まであります。
この数字タグは文字にタイトルや見出しなどの意味を与えます。
ファイルを作成してどのような表示になるか試してみましょう。
htmlフォルダへh.htmlというファイルを作成して中身は以下のようにしてください。

```python
<h1>h1タグです</h1>
<h2>h2タグです</h2>
<h3>h3タグです</h3>
<h4>h4タグです</h4>
<h5>h5タグです</h5>
<h6>h6タグです</h6>
```

保存して、先ほどと同じようにファイルをダブルクリックしてブラウザで表示してみましょう。

このように表示されると成功です。
もし、日本語が文字化けしてしまう場合は

```python
<meta charset="UTF-8">
```

こちらを上の方に記載してみてください。
タグは文字に意味を与えると同時に、簡単な装飾もしてくれます。h数字だと1から６まで徐々に小さくなるようになっていきます。数字が小さいほど大きな見出しになります。
全てのタグが文字を装飾するわけではありませんが、h数字の装飾は特徴的です。

・h数字を使ったよくある例


```python
<h2>はじめに</h2>

<style>
h2 {
    background-color: rgb(210, 242, 245);
    color: rgb(118, 175, 240);
    padding: 5px;
    padding-left: 20px;
}

</style>
```

div

2つ目にして例外的なタグなのですが、divタグは文字に何か意味を与えるという役割はありませんが、中にタグを集めて、まとめる役割があります。
特定のグループを作り、部分的にcssで装飾する時に使います。詳しい使い方の解説はもう少し先で行います。

htmlフォルダ内にdiv.htmlというファイルを作成して動きを確かめてみましょう。

```python
no div
no div

<div>div1</div>
<div>div2</div>
```

 div1,div2が縦に並びましたね。divタグで囲んだものは横並びではなく、縦に並びます。なぜこのようになるのか詳しくは後述します。

p

pタグはparagraph(段落)を表します。
何か文章を書きたい時にpタグで囲むと、囲んだ中の文字はひと塊りの文章という意味を与えられます。
htmlフォルダ内にp.htmlを作成しましょう。

```python
<p>pタグの中に書いた文章は1つの段落という意味を与えられます。ブログなどで、たくさんの文章を書く必要のあるサイトでは、文章を段落ごとにpタグで囲みます。</p>
<p>pタグの中に書いた文章は1つの段落という意味を与えられます。ブログなどで、たくさんの文章を書く必要のあるサイトでは、文章を段落ごとにpタグで囲みます。</p>
```

分かりやすいように、ブラウザを少し小さくして表示しています。pタグで囲むと下の余白をつけてくれます。この効果により段落がわかりやすくなります。

span

span自体は意味を持たないタグです。
divと似ていますが、spande囲っても縦並びにはならずに、横並びになります。
htmlフォルダにspan.htmlというファイルを作ってみましょう。

```python
<span>hello</span>
<span>goodby</span>

<div>hello</div>
<div>goodby</div>
```

spanタグは例えば単語中の1文字だけ色を変えたいという場合に使ったりします。

```python
<span>he<span style="color:red;">l</span>lo</span>
```

このように、1文字だけ色が赤色になりました。書き方はまた後で詳しく解説するので、今はこのような使い方ができることだけ覚えておいてください。

<span>例:<span style="color: green;">h<span style="color: black">e<span style="color: orange;">l <span style="color:red;">l<span style="color: purple;">o</span></span></span></span></span></span>

<span>例:<span style="color: green;">h<span style="color: black">e<span style="color: orange;">l <span style="color:red;">l<span style="color: purple;">o</span></span></span></span></span></span>

a

aタグは、他のWebページへ移動するための「リンク」を作るためのタグです。
htmlフォルダへa.htmlというファイルを作成してください。

```python
<a href="span.html">go to span page</a>
```

ここでは、タグの中に「href="span.html"」というものが入っています。このようにタグには付加情報を書くことができます。HTMLでは付加情報の事を属性と言います。「href属性」のような使い方です。
a.htmlをブラウザで表示すると、

このようなリンクが表示されます。このリンクをクリックすると、先ほど作成したspan.htmlが表示されます。

input/label

これまではただ表示するだけでしたが、inputタグを使うと、ユーザーが入力する部品を作ることができます。また、inputとよくセットで使うのがlabelタグです。
ulとliの関係のように、必ずしもセットでなくても良いですが、入力項目に何を入れるべきかユーザーに教えるためにlabelタグはよく使います。
labelタグ単体や、input以外のタグとセットで使うとは滅多にありません。
htmlフォルダへinput.htmlを作成してください。

```python
<label>名前</label>
<input type="text" name="username">
```

保存してブラウザで見てみましょう。

ウェブアプリケーションでは、inputを使って会員登録やブログシステムなどを組み立てていきます。
inputもimgと同じく終了タグはありません。また、type属性とname属性はほとんどの場合必要となるので、必ず書くようにしましょう。
type属性にはtext以外にも、passwordやnumberなど、様々な入力に合わせて変えることができます。

typeをpasswordにすると、入力した文字が見えないようになります。

form

ただ入力するだけなら、inputタグだけで問題ないのですが、入力された値をウェブサーバーへ送信したいときがあります。会員登録やログイン、ブログの投稿や編集などを行うときはウェブサーバーへ入力値を送信します。
この時、送信先をどのように決めるかということを、formタグに記述していきます。
htmlフォルダへform.htmlを作成してください。

```python
<form method="post" action="create">
<label>ブログタイトル</label>
<input type="text" name="blog_title">
<input type="submit" value="送信">
</form>
```

header/main/footer

HTML5(最新のバージョン)では、新しくタグが追加されました。
以前のバージョンに比べて、タグを増やし、文章に適切な意味を与えることがより重要視されてきたのです。
新しく追加されたタグの中でもよく使われるのが、header, main, footerの３つです。
1つのWebページを大雑把に分割すると、ヘッダー、メイン、フッターに分かれます。

これらを作るときはそれぞれ対応するタグで囲みます。
htmlフォルダへmain.htmlを作成してください。

```python
<header>
ヘッダー
</header>

<main>
メイン
</main>

<footer>
フッター
</footer>
```

これらのタグは内部に意味を与えますが、何か装飾されるわけではありません。

html

htmlタグはWebページ全体を囲むためのタグです。
会社の構造を思い出すと、会社のトップである社長に当たります。
木構造では、一番の根っこの部分は１つの要素から始まっていないといけないので、全てのタグを取りまとめる役割としてhtmlタグがあります。

head

headタグはこれまでのものと大きく異なり、Webページの設定を記述するためのタグです。例えば、cssファイルの読み込みや、フォントファイルの読み込み、サイトのタイトルなどの設定を記述します。
headタグの中には、画面に表示されないような設定用のタグを記述していきます。

body

bodyタグはheadとは対照的に、画面に表示される部分を取りまとめる役割があります。
htmlタグが社長なら、headは表に出ない開発事業部で、bodyは表に出る広報や営業部となります。
これまで登場したdivやmain, ulなどのほとんどのタグはbodyタグの中に書いていきます。
